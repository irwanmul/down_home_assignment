```python
import pandas as pd
import gspread
from gspread_dataframe import set_with_dataframe
from google.colab import auth
from google.auth import default
```

```python
# Authenticate and create the service
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)
```

```python
# Load the datasets
agg_profile_views = pd.read_csv('agg_profile_views.csv')
dim_users = pd.read_csv('dim_users.csv')
dim_bad_actors = pd.read_csv('dim_bad_actors.csv')
```

```python
agg_profile_views
```

```python
# Convert date columns to datetime
agg_profile_views['ds'] = pd.to_datetime(agg_profile_views['ds'])
dim_users['created_at'] = pd.to_datetime(dim_users['created_at'],format='mixed', errors='coerce').dt.tz_localize(None)
```

```python
agg_profile_views
```

```python
# Filter new users (last 30 days from the latest date in the agg_profile_views dataset)
latest_date = agg_profile_views['ds'].max()
start_date = latest_date - pd.Timedelta(days=60)
new_users = dim_users[dim_users['created_at'] >= start_date]
```

```python
# Join agg_profile_views with new users to calculate metrics for new users only
new_user_exposures = pd.merge(
    agg_profile_views,
    new_users,
    left_on='viewed_user_id',
    right_on='user_id',
    how='inner'
)
```

```python
new_user_exposures
```

```python
# Filter out bad actors
valid_user_exposures = new_user_exposures[~new_user_exposures['viewed_user_id'].isin(dim_bad_actors['user_id'])]
```

```python
# Calculate Metrics
total_exposure_count = int(new_user_exposures['cnt'].sum())
exposure_rate_by_filter = new_user_exposures.groupby('filter_type')['cnt'].sum().reset_index()
avg_exposures_per_user = float(new_user_exposures.groupby('viewed_user_id')['cnt'].sum().mean())
exposure_distribution_by_gender = new_user_exposures.groupby('gender')['cnt'].sum().reset_index()
```

```python
# Total Exposures (Raw data per viewed user, including ds)
total_exposure_raw = new_user_exposures[['viewed_user_id', 'cnt', 'ds']]
total_exposure_raw.to_csv('total_exposures_raw.csv', index=False)
```

```python
# Average Exposure per User (Raw data per user)
avg_exposure_raw = new_user_exposures.groupby('viewed_user_id')['cnt'].sum().reset_index()
avg_exposure_raw.columns = ['viewed_user_id', 'total_exposures']
avg_exposure_raw.to_csv('avg_exposures_per_user_raw.csv', index=False)
```

```python
# Exposure Distribution by Gender (Raw data including ds)
exposure_distribution_by_gender_raw = new_user_exposures[['gender', 'cnt', 'ds']]
exposure_distribution_by_gender_raw.to_csv('exposure_distribution_by_gender_raw.csv', index=False)
```

```python
# Exposure Rate by Filter Type (Raw data including ds)
exposure_rate_by_filter_raw = new_user_exposures[['filter_type', 'cnt', 'ds']]
exposure_rate_by_filter_raw.to_csv('exposure_rate_by_filter_raw.csv', index=False)
```

```python
# Calculate the total exposure count with and without bad actors
exposure_count_with_bad_actors = new_user_exposures['cnt'].sum()
exposure_count_without_bad_actors = valid_user_exposures['cnt'].sum()

# Create a DataFrame for comparison
impact_analysis_df = pd.DataFrame({
    'Exposure Type': ['With Bad Actors', 'Without Bad Actors'],
    'Total Exposures': [exposure_count_with_bad_actors, exposure_count_without_bad_actors]
})

# Save the comparison data to CSV
impact_analysis_df.to_csv('impact_analysis_exposure.csv', index=False)
```

```python
# Create Cohort Data (based on user creation week)
new_user_exposures['created_week'] = new_user_exposures['created_at'].dt.to_period('W')

cohort_filter_data = new_user_exposures.groupby(['created_week', 'filter_type'])['cnt'].sum().reset_index()
cohort_filter_data.columns = ['cohort_week', 'filter_type', 'total_exposures']

# Save cohort data to CSV
cohort_filter_data.to_csv('cohort_filter_exposure.csv', index=False)
```
